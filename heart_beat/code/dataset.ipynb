```python
%matplotlib inline
import librosa
import numpy as np
import os
import sys
import re
import argparse
import pathlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#nsongs=20              #the number of audio songs
#sr=44100               #help='sampling rate
n_fft=2048             #number of samples per frame
hpr=0.25                  #the rate of hopping samples per frame
inpath='/media/alvin/disk_D/dataset/HeartBeat' #the path of training files
outpath=''             #the path of output data
HBdb = pd.DataFrame(columns=['sid','dbset','dbtype','fname','noise','dur','sr','wav_raw'])
HBdb['dur'] = HBdb['dur'].astype(float)
```

```python
medset_pascal  = ['/Atrain/Atraining_normal/',
           '/Atrain/Atraining_murmur/',
           '/Atrain/Atraining_extrahls/',
           '/Atrain/Atraining_artifact/',
           '/Atrain/Aunlabelledtest/',
           '/Btrain/Training B Normal/',
           '/Btrain/Btraining_murmur/',
           '/Btrain/Btraining_extrastole/',
           '/Btrain/Bunlabelledtest/',
           '/Btrain/Btraining_noisynormal/',
           '/Btrain/Btraining_noisymurmur/']

medset_physionet  = ['/training/training-a/',
                     '/training/training-b/',
                     '/training/training-c/',
                     '/training/training-d/',
                     '/training/training-e/',
                     '/training/training-f/',
                     '/validation/']
```

```python
# 要處理的wave路徑，唯一要改的是這裡
medset = medset_pascal
#medset = medset_physionet

if medset == medset_pascal:
    inpath += '/pascal'
    o_pd_name = 'HBdb_pascal'
else:
    inpath += '/physionet'
    o_pd_name = 'HBdb_physionet'
```

```python

for medPath in medset:
    thisP    = inpath + medPath
    filelist = os.listdir(thisP)
    filelist = filter(lambda x: x.endswith('.wav'), filelist)
    filelist = list(filelist)
    filelist.sort(key=lambda x : int(re.sub('\D','',x)))
    filelist = list(filelist)
    print('Total traing file in ('+medPath+')= ' + str(len(filelist)))

    if re.search('/',re.sub('^/|/$','',medPath)):
        dbset_name,dbtype_name  = re.sub('^/|/$','',medPath).split('/')
    else:
        dbset_name = re.sub('^/|/$','',medPath)
        dbtype_name = ''
        
    noise_flag = 1 if re.search('noisy',dbtype_name) else 0 
    
    for fname in filelist:
        audio_raw_data, sr = librosa.load(inpath + medPath + fname,sr=None,mono=True) # default mono=True
        print('fname='+fname+'\ttype='+dbtype_name+'\tsr='+str(sr)+'\tdur='+repr(round(len(audio_raw_data)/sr,2)))
        new_sid = HBdb.shape[0]+1
        HBdb.set_value(new_sid,'sid',new_sid)
        HBdb.set_value(new_sid,'dbset',dbset_name)
        HBdb.set_value(new_sid,'dbtype',dbtype_name)
        HBdb.set_value(new_sid,'fname',fname)
        HBdb.set_value(new_sid,'noise',noise_flag)
        HBdb.set_value(new_sid,'dur',len(audio_raw_data)/sr)
        HBdb.set_value(new_sid,'sr',sr)
        HBdb.set_value(new_sid,'wav_raw',audio_raw_data)
    
    
    
```
