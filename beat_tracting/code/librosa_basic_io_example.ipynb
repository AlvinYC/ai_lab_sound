```python
%matplotlib inline
import librosa
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt
#import ipd
```

```python
y, sr = librosa.load(librosa.util.example_audio_file())
```

```python
print(y.shape)
print(sr)
```

# plot all wave

```python
plt.figure(figsize=(15,3))
plt.subplot(1, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('Monophonic')
```

```python
ipd.Audio(y, rate=sr)
```

# plot a part of duration  (method 1) - load file w/ specific duration

```python
y, sr = librosa.load(librosa.util.example_audio_file(),duration=0.5)
```

```python
plt.figure(figsize=(15,3))
plt.subplot(1, 1, 1)
librosa.display.waveplot(y, sr=sr)
plt.title('Monophonic')
```

# plot specific time wave

```python
t_beg = 0.2;
t_dur = 0.02
plt.figure(figsize=(15,3))
plt.subplot(1, 1, 1)
librosa.display.waveplot(y[int(t_beg*sr):int((t_beg+t_dur)*sr)], sr=sr,alpha=0.1)
plt.title('Monophonic')
```

## how many sample in this duration

```python
sample_num = t_dur * sr
print(sample_num)
```

```python
y_harm, y_perc = librosa.effects.hpss(y)
plt.subplot(3, 1, 3)
librosa.display.waveplot(y_harm, sr=sr, alpha=0.25)
librosa.display.waveplot(y_perc, sr=sr, color='r', alpha=0.5)
plt.title('Harmonic + Percussive')
plt.tight_layout()
```
